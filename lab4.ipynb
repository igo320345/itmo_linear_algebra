1. Система асимптотически устойчива

```python
x1 = exp(A1 * t) * x01
x2 = exp(A1 * t) * x02
x3 = exp(A1 * t) * x03

t_vals = np.linspace(0, 10, 1000)

X1, X2, X3, Y1, Y2, Y3 = [], [], [], [], [], []
for t_val in t_vals:
    X1.append(N(x1[0].subs(t, t_val)))
    Y1.append(N(x1[1].subs(t, t_val)))
    X2.append(N(x2[0].subs(t, t_val)))
    Y2.append(N(x2[1].subs(t, t_val)))
    X3.append(N(x3[0].subs(t, t_val)))
    Y3.append(N(x3[1].subs(t, t_val)))
plt.plot(X1, Y1)
plt.plot(X2, Y2)
plt.plot(X3, Y3)
plt.xlabel('x1')
plt.ylabel('x2')
plt.grid(True)
```

```python
plt.plot(t_vals, X1)
plt.plot(t_vals, X2)
plt.plot(t_vals, X3)
plt.xlabel('t')
plt.ylabel('x1')
plt.grid(True)
```
