```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
```

Исходные значения x и y

```python
x_old = np.linspace(-4, 4, 9)
y_old = np.linspace(-3, 3, 7)
xy_old = np.column_stack([[x, y] for x in x_old for y in y_old])
```

Функции для визуализации линейных отображений

```python
def colorizer(x, y):
        r = min(1, 1-y/3)
        g = min(1, 1+y/3)
        b = 1/4 + x/16
        return (r, g, b)

colors = list(map(colorizer, xy_old[0], xy_old[1]))

def plot_grid(xy_grid):
    plt.figure(figsize=(4, 4), facecolor="w")
    plt.scatter(xy_grid[0], xy_grid[1], s = 36, c = colors, edgecolors = 'none')
    plt.axis('equal')
    plt.grid(True)
```

Исходная сетка значений

```python
plot_grid(xy_old)
```

# Задание 1

Выбранные мной четыре числа

```python
a = 3
b = 2
c = 5
d = 4
```

1. Отражение плоскости относительно прямой y = ax

```python
a1 = Matrix([[(1 - a ** 2) / (a ** 2 + 1), 2 * a / (a ** 2 + 1)], [2 * a / (a ** 2 + 1), (a ** 2 - 1) / (a ** 2 + 1)]])
xy_new = np.dot(a1, xy_old)
plot_grid(xy_new)
print(a1)
```

2. Отображение всей плоскости в прямую y = bx

```python
a2 = Matrix([[1, 2], [b, 2 * b]])
xy_new = np.dot(a2, xy_old)
plot_grid(xy_new)
print(a2)
```

```python
a3 = Matrix([[cos(10 * c), sin(10 * c)], [-sin(10 * c), cos(10 * c)]])
xy_new = np.dot(a3, xy_old)
plot_grid(xy_new)
print(a3)
```

4. Центральная симметрия плоскости относительно начала координат

```python
a4 = Matrix([[-1, 0], [0, -1]])
xy_new = np.dot(a4, xy_old)
plot_grid(xy_new)
print(a4)
```

```python
a5 = Matrix([[cos(10 * d), -sin(10 * d)], [sin(10 * d), cos(10 * d)]]) * a1
xy_new = np.dot(a5, xy_old)
plot_grid(xy_new)
print(a5)
```

6. Отображение, которое переводит прямую y = 0 в y = ax и прямую x = 0 в y = bx

```python
a6 = Matrix([[1, a / b], [a, a]])
xy_new = np.dot(a6, xy_old)
plot_grid(xy_new)
print(a6)
```

```python
a7 = a6.inv()
xy_new = np.dot(a7, xy_old)
plot_grid(xy_new)
print(a7)
```
